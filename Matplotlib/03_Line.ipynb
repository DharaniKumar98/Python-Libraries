## Visualisation in Python - Matplotlib

### Line Chart: Trend of sales over the 12 months

-  Can be used to present the trend with time variable on the x-axis
-  In some cases, can be used as an alternative to scatterplot to understand the relationship between 2 variables

- A line graph is very helpful when you want to understand the trend of a variable. Some key industries and services that rely on line graphs are financial markets, weather forecast, etc.

- The syntax for plotting a line graph is plt.plot(x_component, y-component, 'graph_features'). You have to specify the marker as a separate attribute to plot a line graph.

- A line plot and a scatterplot essentially plot the x-y relationship between two quantitative variables. They show how one variable changes with respect to another quantitative variable. To find a relationship using a line plot, one additional step that you have to complete is to first sort the elements of x-axis.

```python
# importing the required libraries
import numpy as np
import matplotlib.pyplot as plt

# Sales data across months
months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
sales = np.array([241268.56, 184837.36, 263100.77, 242771.86, 288401.05, 401814.06, 258705.68, 456619.94, 481157.24, 422766.63, 555279.03, 503143.69])
```

```python
# plotting a line chart
#plt.plot(x_axis, y_axis)
plt.plot(months , sales)
# adding title to the chart
plt.title('Sales Trend across different months of the year'.title() ,fontdict={'fontsize':15 , 'fontweight': 9 , 'color' : 'green'})

# labeling the axes
plt.xlabel('Month', fontdict={'fontsize': 9,
           'fontweight': 7, 'color': 'red'})

plt.ylabel('Sales', fontdict={'fontsize': 9,
           'fontweight': 7, 'color': 'red'})
# rotating the tick values of x-axis
plt.xticks(rotation = 90)

# displaying the created plot
plt.show()
```

- You should be careful when using the function plt.plot. The same function generates a scatterplot if you tweak the syntax and specify the markers.

```python
y = np.random.randint(1, 100, 50)
plt.plot(y, 'r*')  # ‘ro’ represents color (r) and marker (o) /(*)
```

```python
#If you specify the colour and marker separately, then you will get a line plot with the points marked.
plt.plot(y, 'red', marker='o')
```

- A line graph is used to present continuous time-dependent data. It accurately depicts the trend of a variable over a specified time period.

```python
plt.plot(months, sales, 'b', marker='x')
plt.show()
```
